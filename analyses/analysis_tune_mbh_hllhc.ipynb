# Tune with  $b_2$ errors from MBH magnets

```python
import pandas as pd
from madx.twiss import twiss
import matplotlib.pyplot as plt
```

Get the tune for all seeds

```python
basedir = "/media/phermes/local/171018_bbeat/results/"
output = []
for seed in range(1, 61):
    filename = '{0}HLLHC_B1_col_MBH1_MB0_test19_BPM_seed_{1}_ql_1.0000_qr_1.0000.tfs'.format(basedir,seed)
    try:
        t = twiss(filename)
    except:
        continue
    output.append([seed, t.Q1, t.Q2])
output = pd.DataFrame(output,columns=['seed','q1','q2'])
```

```python
plt.plot(output['q1']-62.31, output['q2']-60.32,'o')
plt.xlabel(r"$Q_x-62.31$")
plt.ylabel(r"$Q_y-60.32$")
plt.title("Beam 1")
plt.show()
```
